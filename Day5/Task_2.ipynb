```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.svm import SVC
```

```python
iris = load_iris()
iris
```

```python
xds = iris["data"]
yds= iris["target"]
```

```python
xds
```

```python
yds
```

```python
fnames = iris["feature_names"]
fnames
```

```python
import pandas as pd
dfds=pd.DataFrame(xds, columns=fnames)
dfds
```

```python
X_train, X_test, y_train, y_test = train_test_split(xds, yds, test_size=0.2, random_state=40)
```

```python
# Scaling the features X 
sc = RobustScaler()
X_train= sc.fit_transform(X_train)
X_test= sc.transform(X_test)
```

```python
LR = LogisticRegression(penalty='elasticnet' , l1_ratio=0.2, solver='saga')
LR.fit(X_train,y_train)
```

```python
y_pred1= LR.predict(X_test)
accuracy = accuracy_score(y_test, y_pred1)
print(classification_report(y_test, y_pred1))
print(confusion_matrix(y_test, y_pred1))

print("Accuracy:", accuracy)
```

```python
# Linear Kernel
from sklearn.svm import SVC
sv = SVC(kernel='linear')
sv.fit(X_train,y_train)
```

```python
y_pred5= sv.predict(X_test)
print(classification_report(y_test, y_pred5))
print(confusion_matrix(y_test, y_pred5))
```

```python
from sklearn.svm import SVC
sv = SVC(kernel='sigmoid')
sv.fit(X_train,y_train)
```

```python
y_predsig= sv.predict(X_test)
print(classification_report(y_test, y_predsig))
print(confusion_matrix(y_test, y_predsig))
```

```python
from sklearn.svm import SVC
sv = SVC(kernel='poly')
sv.fit(X_train,y_train)
```

```python
y_predpoly= sv.predict(X_test)
print(classification_report(y_test, y_predpoly))
print(confusion_matrix(y_test, y_predpoly))
```

```python
from sklearn.svm import SVC
sv = SVC(kernel='rbf')
sv.fit(X_train,y_train)
```

```python
y_predrbf= sv.predict(X_test)
print(classification_report(y_test, y_predrbf))
print(confusion_matrix(y_test, y_predrbf))
```
